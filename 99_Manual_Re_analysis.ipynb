```python
import os
import time

from pprint import pprint
import pandas as pd
import numpy as np
from rtm_pymmcore.data_structures import SegmentationMethod
```

```python
base_path = "E:\\Alex\\"
experiment_name = "DemoTestnewThread"
path = os.path.join(base_path, experiment_name)
df_acquire = pd.read_parquet("E:\\Alex\\DemoTestnewThread\\0_latest.parquet")
out_path = os.path.join(base_path, f"{experiment_name}_re")
```

```python
%load_ext autoreload
%autoreload 2

from rtm_pymmcore.segmentation.cellpose_v4 import CellposeV4
from rtm_pymmcore.stimulation.base_stimulation import StimWholeFOV
from rtm_pymmcore.tracking.trackpy import TrackerTrackpy
from rtm_pymmcore.feature_extraction.erk_ktr import FE_ErkKtr
from rtm_pymmcore.feature_extraction.simple_fe import SimpleFE


segmentators = [
    SegmentationMethod(
        name="labels",
        segmentation_class=CellposeV4(min_size=50),
        use_channel=0,
        save_tracked=True,
    ),
]

stimulator = StimWholeFOV()
feature_extractor = SimpleFE("labels")
tracker = TrackerTrackpy()
optocheck_fe = None


from rtm_pymmcore.img_processing_pip import ImageProcessingPipeline_postExperiment

pipeline = ImageProcessingPipeline_postExperiment(
    img_storage_path=path,
    out_path=out_path,
    df_acquire=df_acquire,
    segmentators=segmentators,
    feature_extractor=feature_extractor,
    tracker=tracker,
    feature_extractor_optocheck=optocheck_fe,
    use_old_segmentations=True,
    use_old_stim_masks=True,
    n_jobs=4,
)
```
